```python
import pandas as pd

data = pd.read_csv('California_Housing_CitiesAdded.csv')
```

```python
# View the first few rows
data.head()

# Check the column names and data types
data.info()

# Check for missing values
data.isnull().sum()
```

```python
# Select only numeric columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Fill missing values in numeric columns with their median
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].median())

# Fill missing values in numeric columns with their median
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].median())
```

```python
print(data.columns)
```

```python
# Fill missing values in non-numeric columns with a placeholder or mode
data['City'] = data['City'].fillna('Unknown')
```

```python
print(data.shape)
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot for Median Income vs. Median House Value
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Median_Income', y='Median_House_Value', data=data)  # Use the correct column names
plt.title('Median Income vs. Median House Value')
plt.show()

```

```python
# Check for any remaining missing values
data.isnull().sum()
```

```python
# Feature engineering with correct column names
data['Rooms per Household'] = data['Tot_Rooms'] / data['Households']
data['Bedrooms per Room'] = data['Tot_Bedrooms'] / data['Tot_Rooms']
data['Population per Household'] = data['Population'] / data['Households']

```

```python
# Select only numeric columns for correlation matrix
correlation_matrix = data.select_dtypes(include=['float64', 'int64']).corr()

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
print(data.columns)

```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='City', y='Median_House_Value', data=data)
plt.xticks(rotation=90)
plt.title('House Prices by City')
plt.show()
```

```python
X = data[['Median_Income', 'Tot_Rooms', 'Population', 'Households', 'Rooms per Household', 'Bedrooms per Room']]
y = data['Median_House_Value']
```
